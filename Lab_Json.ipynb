```python
import pandas as pd

df = pd.read_csv("data/titanic.csv")
print(df.head())
print(df.shape)
```

```python
import numpy as np

print("\n" + "="*50)
print("DESCRIPTIVE STATISTICS")
print("="*50)

# Select numeric columns
numeric_columns = df.select_dtypes(include=[np.number]).columns

for col in numeric_columns:
    print(f"\nColumn: {col}")
    print(f"Mean: {df[col].mean():.2f}")
    print(f"Median: {df[col].median():.2f}")
    print(f"Std Dev: {df[col].std():.2f}")
```

```python
print("\n" + "="*50)
print("MISSING VALUES ANALYSIS")
print("="*50)

for col in df.columns:
    missing_count = df[col].isnull().sum()
    missing_percent = (missing_count / len(df)) * 100

    print(f"{col}: {missing_count} missing ({missing_percent:.2f}%)")
```

```python
print("\n" + "="*50)
print("FEATURE ENGINEERING")
print("="*50)

# Create copy
df_features = df.copy()

# 1. Family Size
df_features['FamilySize'] = df_features['SibSp'] + df_features['Parch'] + 1

print("\nFamilySize:")
print(df_features[['SibSp', 'Parch', 'FamilySize']].head(10))


# 2. Is Alone
df_features['IsAlone'] = (df_features['FamilySize'] == 1).astype(int)

print("\nIsAlone:")
print(df_features[['FamilySize', 'IsAlone']].head(10))


# 3. Age Group
def categorize_age(age):
    if pd.isna(age):
        return 'Unknown'
    elif age < 18:
        return 'Child'
    elif age < 30:
        return 'Young Adult'
    elif age < 50:
        return 'Adult'
    else:
        return 'Senior'

df_features['AgeGroup'] = df_features['Age'].apply(categorize_age)

print("\nAgeGroup:")
print(df_features[['Age', 'AgeGroup']].head(10))
```

```python
print("\n" + "="*50)
print("FEATURE ANALYSIS: SURVIVED vs NOT SURVIVED")
print("="*50)

# Group comparison
print("\nFamily Size by Survival:")
print(df_features.groupby('Survived')['FamilySize'].mean())

print("\nIs Alone by Survival:")
print(df_features.groupby('Survived')['IsAlone'].mean())

print("\nAge Group vs Survival:")
print(pd.crosstab(df_features['AgeGroup'], df_features['Survived'], normalize='index') * 100)
```

```python
import json
from datetime import datetime

# Passenger class
class Passenger:
    def __init__(self, row):
        self.passenger_id = int(row['PassengerId'])
        self.name = row['Name']
        self.age = None if pd.isna(row['Age']) else float(row['Age'])
        self.sex = row['Sex']
        self.survived = int(row['Survived'])
        self.pclass = int(row['Pclass'])
        self.fare = float(row['Fare'])
        self.family_size = int(row['FamilySize'])
        self.is_alone = int(row['IsAlone'])
        self.age_group = row['AgeGroup']

    def to_dict(self):
        return self.__dict__


# Dataset class
class TitanicDataset:
    def __init__(self, df):
        self.df = df
        self.passengers = [Passenger(row) for _, row in df.iterrows()]

    def get_summary(self):
        return {
            "total_passengers": len(self.passengers),
            "survived": int(self.df['Survived'].sum()),
            "not_survived": int(len(self.df) - self.df['Survived'].sum()),
            "survival_rate": float(self.df['Survived'].mean())
        }

    def to_json(self, filename="data/titanic_processed.json"):
        data = {
            "metadata": {
                "created_at": datetime.now().isoformat(),
                "dataset": "Titanic"
            },
            "summary": self.get_summary(),
            "passengers": [p.to_dict() for p in self.passengers]
        }

        with open(filename, "w") as f:
            json.dump(data, f, indent=2)

        print("JSON file created successfully!")


# Create dataset and export
dataset = TitanicDataset(df_features)
dataset.to_json()
```
